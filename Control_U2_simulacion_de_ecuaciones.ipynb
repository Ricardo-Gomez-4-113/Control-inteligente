## Formulas
\begin{equation}
\ddot{Θ} + \frac{g}{l} \sin{Θ} = 0
\end{equation}
\begin{equation}
Θ = x_1
\end{equation}
\begin{equation}
\dot{Θ} = x_2
\end{equation}
\begin{equation}
\ddot{Θ} = \dot{x_2}
\end{equation}
\begin{equation}
\dot{x_2} = - \frac{g}{l} \sin{Θ}
\end{equation}

```python
# Importamos liberias
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Definimos la funcion
def mydiff(x,t,l,g):
  dx1dt = x[1]
  dx2dt = -(g/l)*np.sin(x[0])
  dxdt = [dx1dt,dx2dt]
  return dxdt

# Definimos el incio, fin de la simulacion y el incremento para la simulacion
tstart = 0
tstop = 5
increment = .1

# Definimos las constantes
g = 9.81
l = 0.3

# Creamos un arreglo con las constantes
t = np.arange(tstart, tstop+increment, increment)
x_init = [np.deg2rad(45),0]
x = odeint(mydiff,x_init,t,args=(l,g,))

# Definimos las variables
x1 = x[:,0]
x2 = x[:,1]

# Graficamos
plt.plot(t,x1,label=r'$\theta$')
plt.plot(t,x2,label=r'$\dot{\theta}$')
plt.legend()
plt.show()
```

## Formulas
\begin{equation}
m\ddot{x} + b\ddot{x} + kx = 0
\end{equation}
\begin{equation}
x = x_1
\end{equation}
\begin{equation}
\dot{x} = x_2
\end{equation}
\begin{equation}
\ddot{x} = \dot{x_2}
\end{equation}
\begin{equation}
\dot{x_1} = x_2
\end{equation}
\begin{equation}
\dot{x_2} = - \frac{b}{m} \dot{x} - \frac{k}{m}x
\end{equation}

```python
# Importamos liberias
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Definimos la funcion
def mydiff(x,t,k,m,b):
  dx1dt = x[1]
  dx2dt = -(b/m*x[1]) - (k/m*x[0]) 
  dxdt = [dx1dt,dx2dt]
  return dxdt

# Definimos el incio, fin de la simulacion y el incremento para la simulacion
tstart = 0
tstop = 20
increment = .1

# Definimos las constantes
k = 5 # constante del resorte (N/m)
m = 10 # masa (kg)
b = 5
# Creamos un arreglo con las constantes
t = np.arange(tstart, tstop+increment, increment)
x_init = [0,1]
x = odeint(mydiff,x_init,t,args=(k,m,b,))

# Definimos las variables
x1 = x[:,0]
x2 = x[:,1]

# Graficamos
plt.plot(t,x1,label=r'$x$')
plt.plot(t,x2,label=r'$\dot{x}$')
plt.legend()
plt.show()
```

# Formulas
\begin{equation}
m\ddot{x} + k_1(x-y) + k_2x = 0
\end{equation}
\begin{equation}
k_1(x-y) = b_1y
\end{equation}
\begin{equation}
x = -\frac{x_1}{m}(x-y)-\frac{k_2}{m}x
\end{equation}
\begin{equation}
y = -\frac{k_1}{b_1}(x-y)
\end{equation}

---
\begin{equation}
\ddot{x} = \dot{x_2} = -\frac{k_1}{m}(x_1-y)-\frac{k_2}{m}x_1
\end{equation}
\begin{equation}
\ddot{x} = x_2
\end{equation}
\begin{equation}
y = x = -\frac{k_1}{b}(x-y)
\end{equation}


---

\begin{equation}
x_1 = x
\end{equation}
\begin{equation}
\dot{x_1} = \dot{x}
\end{equation}
\begin{equation}
x_2 = \dot{x}
\end{equation}
\begin{equation}
\ddot{x} = \dot{x}
\end{equation}
\begin{equation}
\dot{x_1} = x_2
\end{equation}
\begin{equation}
\dot{x_2} = -\frac{k_1}{m}(x_1-x_3) - \frac{k_2}{m}(\dot{x_1})
\end{equation}
\begin{equation}
x_3 = y
\end{equation}
\begin{equation}
\dot{x_3} = \dot{y}
\end{equation}
\begin{equation}
\dot{x_3} = \frac{k_1}{b}(x_1-x_3)
\end{equation}

```python
# Importamos liberias
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Definimos la funcion
def mydiff(x,t,k1,k2,m,b):
  dx1dt = x[1]
  dx2dt = -(k1/m)*(x[0]-x[2])-(k2/m)*x[0]
  dx3dt = (k1/b)*(x[0]-x[2])
  dxdt = [dx1dt,dx2dt,dx3dt]
  return dxdt

# Definimos el incio, fin de la simulacion y el incremento para la simulacion
tstart = 0
tstop = 7
increment = .1

# Definimos las constantes
k1 = 50 # constante del resorte (N/m)
k2 = 31.75 # constante del resorte (N/m)
m = .2 # masa (kg)
b = .4
# Creamos un arreglo con las constantes
t = np.arange(tstart, tstop+increment, increment)
x_init = [1,0,0.5]
x = odeint(mydiff,x_init,t,args=(k1,k2,m,b,))

# Definimos las variables
x1 = x[:,0]
x2 = x[:,1]
x3 = x[:,2]
# Graficamos
plt.plot(t,x1,label=r'$x$')
plt.plot(t,x2,label=r'$\dot{x}$')
plt.plot(t,x3,label=r'y')
plt.legend()
plt.show()
```
